```python
import pandas as pd
df = pd.read_csv('captions.txt', delimiter = ",")
df.columns = ['image_id', 'caption']
df = df.drop_duplicates(subset=['image_id'], keep='first')
print(df.tail())
delete_arr = df['image_id'].tail(3).to_numpy()
print(delete_arr)
import os, glob
for filename in delete_arr:
    os.remove(f'/content/Images/flickr30k_images/{filename}') 
    continue
df = df.iloc[0:-3]
print(df.tail())
print(len(df))
```

```python
import cv2
import os
from google.colab.patches import cv2_imshow
import os

for filename in os.listdir("/content/Images/flickr30k_images"):
  image_cv2 = cv2.imread(f"/content/Images/flickr30k_images/{filename}", cv2.IMREAD_UNCHANGED) 
  img_stretch = cv2.resize(image_cv2, (64, 64))
  cv2.imwrite('/content/images/'+filename, img_stretch)
```
